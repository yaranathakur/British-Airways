```python
import pandas as pd
import numpy as np

import requests

from bs4 import BeautifulSoup

url = 'https://www.airlinequality.com/airline-reviews/british-airways'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'lxml')
tables = soup.find_all('table', class_ = 'review-ratings')
tables
```

```python
for table in tables:
    table_data = {}  # Dictionary to hold the data for the current table
    
    # Iterate over each row in the table
    for tr in table.find_all('tr'):
        tds = tr.find_all('td')
        if len(tds) == 2:  # Ensure the row has exactly two <td> elements
            column_name = tds[0].get_text(strip=True)  # First <td> is the column name
            column_value = tds[1].get_text(strip=True)  # Second <td> is the value
            table_data[column_name] = column_value  # Add to the dictionary
    
    # Convert the dictionary to a DataFrame (single row DataFrame)
    df = pd.DataFrame([table_data])
    df_list.append(df)

# Optionally, combine all the DataFrames into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)
```
